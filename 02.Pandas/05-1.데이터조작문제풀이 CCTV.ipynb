#### 선생님 풀이

```python
cctv = pd.read_csv('data/서울시 자치구 연도별 방범용 CCTV 운영 현황_230630기준.csv', encoding='EUC-KR')
cctv.head(3)
```

```python
# 위의 DF를 구분을 index로 주고 싶어 : index_col='구분'
# 앞에 Unnamed: 0 컬럼을 지워주고 싶어 : del cctv['Unnamed: 0']
cctv = pd.read_csv('data/서울시 자치구 연도별 방범용 CCTV 운영 현황_230630기준.csv', encoding='EUC-KR', index_col='구분')
del cctv['Unnamed: 0']
cctv.head(3)
```

```python
# 각 연도별 컬럼에, 첫번째 데이터의 타입을 확인하겠다!
for col in cctv.columns:
    print(col, type(cctv[col][0]))      
```

- 데이터 조작에서 str를 쓸수 있는것.
    - find
    - replace

```python
for col in cctv.columns:
    if col in ['2021년','2022년']:
        continue                    # 2021년, 2022년의 경우, 변형없이 그냥 가고
    cctv[col] = cctv[col].str.replace(',', '').astype(int)
```

```python
cctv['계'] = cctv.sum(axis=1)
cctv.head()
```

```python
# 더 간단한 다른 선생님 풀이방법
cctv = pd.read_csv('data/서울시 자치구 연도별 방범용 CCTV 운영 현황_230630기준.csv', 
                   encoding='EUC-KR', index_col='구분', thousands=',')
del cctv['Unnamed: 0']
cctv.head()
```
