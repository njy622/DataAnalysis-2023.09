```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.tail(3)
```

```python
# 20살 미만이면 성별에 관계없이 “미성년자”라고 한다.
# 20살 이상이면 나이에 따라 “청년”, “중년”, “장년”, “노년”을 구분하고 그 뒤에 성별을 나타내는 “남성”, “여성”을 붙인다.

bins = [0, 19, 30, 50, 70, 100]             # 1 < 미성년자 <= 20
labels = ["미성년자", "청년", "중년", "장년", "노년"]
titanic['age_group'] = pd.cut(titanic.age, bins, labels=labels)
titanic.tail()
```

```python
titanic['category3'] = \
    titanic.apply(lambda x: x.age_group + '남성' if (x.age >= 20) & (x.sex == 'male') else \
                            x.age_group + '여성' if (x.age >= 20) & (x.sex == 'female') else \
                                '미성년자' if x.age < 20 else np.nan, axis=1)
titanic.head(10)
```

```python
# age의 결측치를 평균으로 대체
titanic.age.fillna(titanic.age.mean().round(2), inplace=True)
titanic['age_group'] = pd.cut(titanic.age, bins, labels=labels)
titanic.tail()
```

```python
titanic['성별'] = titanic.sex.apply(lambda x: '남성' if x == 'male' else '여성')
titanic.tail()
```

```python
titanic['category3'] = titanic.apply(
    lambda x: x.age_group + x.성별 if x.age >= 20 else '미성년자', axis=1)
titanic.tail(10)
```

```python
#학생 이름을 나타내는 열을 포함시키지 않고 데이터프레임 df_score1 을 
# 생성한 후, df_score1.index 속성에 학생 이름을 나타내는 열을 지정하여 
# 인덱스를 지정한다.  reset_index 명령으로 이 인덱스 열을 명령으로 
# 일반 데이터열로 바꾸여 데이터프레임 df_score2을 만든다.
#학생 이름을 나타내는 열이 일반 데이터 열을 포함하는 데이터프레임 
#df_score2에 set_index 명령을 적용하여 다시 학생 이름을 나타내는 열을
# 인덱스로 변경한다.

df_score1 = pd.DataFrame(np.random.randint(60, 101, size=(5, 3)),
                         columns=['국어', '영어', '수학'])
df_score1
```

```python
df_score1.index = ['제임스', '마리아', '파이썬', '김자바', '최다비']
df_score1.index.name = '이름'
df_score1
```

```python
df_score2 = df_score1.reset_index()
df_score2
```

```python
df_score2.set_index('이름', inplace=True)
df_score2
```
