## 선생님 풀이
### 1. CCTV데이터

```python
# CCTV 갯수 : 2015년 ~ 2022년 계
cctv = pd.read_csv('data/서울시 자치구 연도별 방범용 CCTV 운영 현황_230630기준.csv', 
            encoding='EUC-KR', index_col='구분', thousands=',')

cctv.head(3)
```

```python
cctv.drop(columns=['Unnamed: 0', '2023년'], inplace=True)
cctv.head(3)
```

```python
cctv['CCTV댓수'] = cctv.sum(axis=1)
cctv.head(3)
```

```python
cctv['최근증가율'] = ((cctv['2020년']+ cctv['2021년'] + cctv['2022년']) / cctv['CCTV댓수'] * 100).round(2)
cctv.head(3)
```

```python
## 위의 방법을 슬라이싱방식으로 
cctv['최근증가율'] = (cctv.iloc[:, 5:-2].sum(axis=1) / cctv['CCTV댓수'] *100).round(2)
cctv.head(3)
```

```python
cctv.drop(columns='2015년	2016년	2017년	2018년	2019년	2020년	2021년	2022년'.split(), inplace=True)
cctv.head(3)
```

### 2. 서울시 인구데이터 정리하기

```python
person = pd.read_csv('data/서울시 구별 주민등록인구.csv',
                    thousands=',',  skiprows=1 )
person.drop(columns=['세대당인구 (명)','동별(1)','세대 (세대)','세대 (세대).1','계 (명).1','한국인 (명).1','등록외국인 (명).1','인구밀도'], inplace=True)

person
```

```python
person.columns = ['구별', '인구수', '내국인', '외국인','고령자']
person
```

```python
# 첫번째 행 지우기
person.drop([0], inplace=True)
person
```

```python
# 구별을 인덱스로 만들기
person.set_index('구별', inplace=True)
person
```

```python
# 외국인 비율, 고령자 비율
person['외국인 비율'] = (person.외국인/person['인구수']*100).round(2)
person['고령자 비율'] = (person['고령자']/person['인구수']*100).round(2)
person
```

### 3.CCTV 데이터와 POP 데이터 합치기

```python
df = pd.merge(cctv, person, left_index=True, right_index=True)
df
```
