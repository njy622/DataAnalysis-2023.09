#### 1. 피벗(pivot) 테이블
- '피벗(Pivot)`은 몸의 중심축을 한 발에서 다른 발로 옮긴다는 뜻

```python
data = {
    "도시": ["서울", "서울", "서울", "부산", "부산", "부산", "인천", "인천"],
    "연도": ["2015", "2010", "2005", "2015", "2010", "2005", "2015", "2010"],
    "인구": [9904312, 9631482, 9762546, 3448737, 3393191, 3512547, 2890451, 263203],
    "지역": ["수도권", "수도권", "수도권", "경상권", "경상권", "경상권", "수도권", "수도권"]
}
columns = ["도시", "연도", "인구", "지역"]
df1 = pd.DataFrame(data, columns=columns)
df1
```

```python
df1.pivot(values='인구', index='도시', columns='연도')
```

```python
df1.pivot(values='인구', index=['지역','도시'], columns='연도')
# values= , index= , columns= 같은 키워드를 써야 사용가능
```

```python
df1.pivot_table('인구', ['지역', '도시'], '연도')
# values= , index= , columns= 같은 키워드를 안써도 사용가능
```

### 2.그룹 분석

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
```

- group by

```python
iris.describe()
```

```python
iris.species.value_counts()
```

- 칠판메모
    - 오라클 DB에서 아래와 같이 한것처럼 데이터프레임에서도 유사
        - SElECT 집계함수  GROUP BY (  ) HAVING 조건

```python
# Iris 품종별 항목의 갯수
iris.groupby('species').count()     #결과는 데이터프레임
```

```python
# Iris 품종별(species) feature의 평균
iris.groupby('species').mean()      #결과는 데이터프레임
```

```python
# 품종별 sepal(꽃받침) length(길이)의 평균
iris.groupby('species').mean()['sepal_length']      # 결과는 Series(시리즈)
```

```python
iris.groupby('species')['sepal_length'].mean()      #위와 동일한 결과
```

```python
# 품종별 sepal length(꽃받침 길이)의 평균을 데이터프레임으로
iris[['sepal_length', 'species']].groupby('species').mean()
```

![Alt text](image.png)

```python
# 품종별 피쳐의 평균과 표준편차
iris.groupby('species').agg(['mean', 'std'])
# .agg([함수명]) = 원하는 함수명을 작성해주면, 알아서 함수값을 구해줌
# aggregate의 약자: agg (종합하다, 집합하다;모이다)
```

#### agg() 메소드 ?
- agg() 메서드는 다중집계작업(multiple aggregation operation)을 간결하게 나타낼수 있도록 한다.
- 한번의 메서드에 여러 함수를 동시에 입력하고 결과를 출력할수 있다.

```python
# sepal length의 품종별 평균, 표준편차, 최대값, 최소값
iris.groupby('species')['sepal_length'].agg(['mean', 'std', 'max'])
# mean: 평균
# std: 표준편차 (standard deviation)
# max: 최댓값
```

```python
# tip_pct 컬럼 생성 (tip의 퍼센트)
tips['tip_pct'] = (tips.tip / tips.total_bill * 100).round(2)
tips.tail()
```

```python
# 성별 팁비율의 평균, 최대, 최소
tips.groupby('sex')['tip_pct'].agg(['mean', 'max', 'min'])
```

```python
# 흡연 유무별로 팁비율의 평균, 표준편차, 최소, 최대
tips.groupby('smoker')['tip_pct'].agg(['mean', 'std', 'min', 'max'])
```

```python
# 성별, 흡연유무별 팁비율의 평균, 표준편차
tips.groupby(['sex', 'smoker'])['tip_pct'].agg(['mean', 'std'])
# 여러그룹으로 하고 싶으면 리스트로 표현!
```

```python
# 요일별, 시간대별 팁 비율의 횟수, 평균, 표준편차
tips.groupby(['day', 'time'])['tip_pct'].agg(['count', 'mean', 'std'])
```

#### 3. 피벗 테이블(pivot_table)

- pivot table의 순서
    - 밸류, 인덱스, 컬럼, 함수

```python
# 성별, 흡연유무별 팁의 비율
tips.pivot_table('tip_pct', 'sex', 'smoker', 'mean')
```

```python
# 합계 기능
# 성별, 흡연유무별 인원
tips.pivot_table('tip_pct', 'sex', 'smoker', 'count', margins=True, margins_name='계')
```

#### 4. groupby vs pivot_table

```python
# 성별 팁 비율의 평균
tips.groupby('sex')[['tip_pct']].mean()
```

```python
tips.pivot_table('tip_pct', 'sex', aggfunc='mean')  
#컬럼 건너뛰고 함수 바로 줄때 aggfunc= 사용
```

```python
# aggfunc 생략하면 mean
tips.pivot_table('tip_pct', 'sex')
```

- 흡연 유무에 따른 평균 팁 비율과 최소값, 최대값

```python
tips.groupby('smoker')[['tip_pct']].agg(['mean', 'min', 'max'])
```

```python
tips.pivot_table('tip_pct', 'smoker', aggfunc=['mean', 'min', 'max'])
```

![Alt text](image-2.png)

```python
# 팁의 비율이 요일과 점심/저녁 여부, 인원수에 어떤 영향을 받는지 살펴본다.
tips.pivot_table('tip_pct', 'time', aggfunc='count')
```

```python
# 어떤 요인이 가장 크게 작용하는지 판단할 수 있는 방법이 있는가?
# 1) 최소값 - 최대값의 차이가 큰 경우
# 2) 표준편차의 차이가 큰경우
# 요일별
tips.groupby('day')['tip_pct'].agg(['mean','std'])
tips.pivot_table('tip_pct', 'day', aggfunc=['mean', 'std'])
```

```python
# 시간대별
tips.groupby('time')['tip_pct'].agg(['mean','std'])
tips.pivot_table('tip_pct', 'time', aggfunc=['mean', 'std'])
```

```python
# 인원수
tips.groupby('size')['tip_pct'].agg(['mean','std'])
tips.pivot_table('tip_pct', 'size', aggfunc=['mean', 'std'])
```

    - qcut 명령으로 세 개의 나이 그룹을 만든다.

    - 성별, 선실, 나이 그룹에 의한 생존율을 데이터프레임으로 계산한다. 행에는 성별 및 나이 그룹에 대한 다중 인덱스를 사용하고 열에는 선실 인덱스를 사용한다. 생존률은 해당 그룹의 생존 인원수를 전체 인원수로 나눈 값이다.

    - 성별 및 선실에 의한 생존율을 피봇 데이터 형태로 만든다.

```python
# qcut 명령으로 세개의 나이 그룹을 만든다
titanic['age_group'] = pd.qcut(titanic.age, 3, labels=['A1', 'A2', 'A3'])
```

```python
# 2. 성별, 선실, 나이 그룹에 의한 생존율을 데이터 프레임으로 계산한다.
titanic.pivot_table('survived', ['sex', 'class'], 'age_group')
```

```python
# 3. 성별 및 선실에 의한 생존율
titanic.pivot_table('survived', 'sex', 'class')
```
