## 인구소멸 위기지역 - 키토그램 시각화(Part 3)

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False # x,y축 레이블 - 깨지는 현상방지
```

```python
pop = pd.read_csv('data/시군구_ID부여.csv', encoding='euc-kr')
pop.head()
```

- 지도 모양 보기

```python
map_raw = pd.read_excel('data/draw_korea_raw(2021).xlsx')
map_raw
```

```python
# table 모양을 변경
map_raw.stack()[:10]
```

```python
map = pd.DataFrame(map_raw.stack())
map.reset_index(inplace=True)
print(map.shape)
map.head()
```

```python
map.columns = ['y', 'x', 'ID']
map.head()
```

- 광역시 경계를 포함하여 그리기

```python
from draw_map import BORDER_LINES, drawKorea, drawKoreaMinus
```

```python
plt.figure(figsize=(8,11))

for idx, row in map.iterrows():
    # 광역시는 구 이름이 겹치는 경우가 많아서 광역시 이름도 같이 표시 
    # (중구, 서구 등)
    if len(row.ID.split())==2:
        dispname = f'{row.ID.split()[0]}\n{row.ID.split()[1]}'
    elif row.ID[:2]=='고성':
        dispname = '고성'
    else:
        dispname = row.ID

    # 서대문구, 서귀포시 같이 이름이 3자 이상인 경우에 작은 글자로 표시
    if len(dispname.splitlines()[-1]) >= 3:
        fontsize, linespacing = 9.5, 1.5
    else:
        fontsize, linespacing = 11, 1.2

    plt.annotate(dispname, (row['x']+0.5, row['y']+0.5), weight='bold',
                 fontsize=fontsize, ha='center', va='center', 
                 linespacing=linespacing)
    
# 시도 경계
for path in BORDER_LINES:
    ys, xs = zip(*path)
    plt.plot(xs, ys, c='black', lw=1.5)

plt.gca().invert_yaxis()
#plt.gca().set_aspect(1)

plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
pop= pd.merge(pop, map)
pop.head()
```

#### 인구 현황 및 인구소멸 위기지역 확인
- 시군구별 인구

```python
drawKorea('인구수계', pop, 'Blues')
```

- 인구소멸 위기 비율

```python
drawKorea('소멸비율', pop, 'Reds')
```

- 인구소멸 위기 지역

```python
pop.소멸위기지역 = pop.소멸위기지역.astype(int)
drawKorea('소멸위기지역', pop, 'Reds')
```

- 인구소멸 고위험지역

```python
pop.소멸위기고위험지역 = pop.소멸위기고위험지역.astype(int)
drawKorea('소멸위기고위험지역', pop, 'Greens')
```

- 20 ~ 30대 성비율

```python
pop['20-30대 성비율'] = pop['20~39세여자'] / pop['20~39세남자'] -1 # 남자가 많으면 +, 여자가 많으면 -
drawKoreaMinus('20-30대 성비율', pop, 'RdBu_r')
```
