## 인구소멸 위기지역 - 전처리(Part 1)

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data/시군구_성_연령_5세_별_인구2_2020.csv', encoding='EUC-KR')
df.shape
```

```python
df.head(260).tail()
```

```python
df.head(10)
```

```python
# 행정구가 있으면 시군구 값을 행정구 값으로 변경
df.시군구[df.행정구.notnull()] = df.행정구[df.행정구.notnull()]
df.tail()
```

```python
# 소계 데이터 지우기 (필터링)
df = df[df.시군구 != '소계']
df.head()
```

```python
# 20~39세, 65세이상 컬럼 만들기
df['20~39세'] = df['20 - 24세']+df['25 - 29세']+df['30 - 34세']+df['35 - 39세']
df['65세 이상'] = df['65 - 69세'] + df['70 - 74세'] + df['75 - 79세'] + df['80세 이상']
```

```python
# 필요한 컬럼 추출
df = df[['광역시도', '시군구', '구분', '인구수', '20~39세','65세 이상']]
print(df.shape)
df.head()
```

```python
# 테이블 정리
pop = pd.pivot_table(df, 
                     index=['광역시도', '시군구'],
                    columns=['구분'],
                    values=['인구수', '20~39세', '65세 이상'] )
pop.head()
```

```python
# 인구 소멸비율 계산
pop['소멸비율'] = round(pop['20~39세', '여자']/ pop['65세 이상', '계'], 4)  # 소숫점 넷째자리까지
```

```python
# 소멸위기지역, 소멸위기 고위험지역
pop['소멸위기지역'] = pop['소멸비율'] < 0.5
pop['소멸위기고위험지역'] = pop['소멸비율'] < 0.2
pop.head()
```

```python
# 인구소멸 위기 지역
crisis_region = pop[pop.소멸위기지역].index.get_level_values(1)
len(crisis_region), crisis_region[:5]
```

```python
# 인구소멸 고위험지역
hiegh_crisis_region = pop[pop.소멸위기고위험지역].index.get_level_values(1)
len(hiegh_crisis_region), hiegh_crisis_region[:5]
```

```python
# 컬럼명 변경: 이중 인덱스 → 단일 인덱스
pop.columns.get_level_values(0)
```

```python
pop.columns.get_level_values(1)
```

```python
new_columns = []
for i in range(len(pop.columns.get_level_values(0))):
    new_columns.append(pop.columns.get_level_values(0)[i]+pop.columns.get_level_values(1)[i])
pop.columns = new_columns
pop.head(3)
```

```python
# 65세 이상남자, 65세 이상여자, 인구수남자, 인구수여자 컬럼 제거
pop.drop(columns=['65세 이상남자', '65세 이상여자', '인구수남자', '인구수여자'], inplace=True)
pop.head(3)
```

```python
# 인덱스 제거
pop.reset_index(inplace=True)
pop.head(3)
```

```python
# 시군구 공백 제거
pop.시군구 = pop.시군구.str.strip()
```

```python
# 파일 저장
pop.to_csv('data/시군구_전처리완료.csv', encoding='euc-kr', index=False)
```
