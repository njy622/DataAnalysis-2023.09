```python
content = div.select_one('.news_end.font1.size3')
answer = ''
for c in content:
    text = c.get_text().strip()
    if text == '':
        continue
    answer += text + '\n'
print(answer)
```

```python
# 세부페이지 데이터 읽기
driver = webdriver.Chrome()
last_page = int(soup.select_one('.paginate').get_text().strip().split('\n')[-1])
for page in range(1, last_page):
    url = f'https://sports.news.naver.com/kbaseball/news/index?isphoto=N&page={page}'
    driver.get(url)
    time.sleep(2)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    lis = soup.select('.news_list > ul > li')
    for _ in range(1,21):
        answer_num = driver.find_element(By.CSS_SELECTOR, '.title').click()
        sub_url = li.select_one('a.thmb')['href']
        sub_url = f'https://sports.news.naver.com/'+sub_url
        sub_soup = BeautifulSoup(driver.page_source, 'html.parser')
        divs = sub_soup.select('.content_area')
    
        answer_list = []
        for div in divs:
            title = div.select_one('.title').get_text().strip()
            content = div.select_one('.news_end font1 size3')
            answer = ''
            for c in content:
                text = c.get_text().strip()
                if text == '':
                    continue
                answer += text + '\n'
            answer_list.append(answer + '\n')
        driver.back()
```
