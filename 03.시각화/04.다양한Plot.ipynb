### 다양한 플롯

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False # x,y축 레이블 - 깨지는 현상방지
```

### 2. Bar Plot

```python
plt.figure(figsize=(5,3))
plt.bar([0,1,2],[2,3,1])
plt.title('Bar Chart')
plt.xticks([0,1,2], ['가', '나', '다']), plt.yticks([1,2,3])
plt.xlabel('가나다'), plt.ylabel('빈도수')
plt.show()
```

```python
s = pd.Series(np.random.rand(10), index=list('abcdefghij'))
plt.bar(s.index, s.values)
plt.title('Bar chart')
plt.show()
```

```python
s.plot(kind='bar'), plt.show()
```

```python
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan']
plt.bar(s.index, s.values, color=colors)
plt.title('Bar chart')
plt.show()
```

```python
plt.barh(s.index, s.values, color=colors)       #barh 로하면 그래프 눕힘가능
plt.title('Horizontal Bar chart')
plt.show()
```

```python
s.index = '종로구 중구 용산구 성동구 광진구 동대문구 중랑구 성북구 강북구 도봉구'.split()
plt.barh(s.index, s.values, color=colors)       #barh 로하면 그래프 눕힘가능
plt.title('Horizontal Bar chart')
plt.show()
```

### 3. 파이차트

```python
ratio = [10, 15, 20, 28]  # 합이 100이 아니어도 됨
labels = ['감', '귤', '밤', '배']
plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.title('파이차트')
plt.show()
```

```python
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.pie(ratio,  labels=labels, autopct='%1.1f%%', colors=colors, explode=explode,
         shadow=True, startangle=90)
plt.axis('equal')
plt.title("Pie Chart")
plt.show()
```

### 5. 산점도(Scatter plot)

```python
import seaborn as sns
iris = sns.load_dataset('iris')
X, Y = iris.sepal_length, iris.sepal_width
plt.scatter(X,Y)
plt.title('산점도 - Iris sepal length vs width')
plt.show()
```

```python
colors = ['red'] * 50 + ['green'] * 50 + ['blue'] * 50
plt.scatter(X, Y, color=colors)
plt.title('산점도 - Iris sepal length vs width')
plt.show()
```

```python
size = 500 * np.random.rand(150)
plt.scatter(X, Y, c=colors, alpha=0.5, s=size)
plt.title('Bubble chart')
plt.show()
```

### 7. Box Plot

```python
# setosa 품종의 각 피처(ssl, ssw, spl, spw)  #필터링할때 이런식으로 씀 iris[iris.species == 'setosa']
ssl = iris[iris.species == 'setosa'].sepal_length
ssw = iris[iris.species == 'setosa'].sepal_width
spl = iris[iris.species == 'setosa'].petal_length
spw = iris[iris.species == 'setosa'].petal_width
```

```python
plt.boxplot((ssl, ssw, spl, spw),
            labels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
plt.title('Setosa 품종의 Feature')
plt.show()
```
