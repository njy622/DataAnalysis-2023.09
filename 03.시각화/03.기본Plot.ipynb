### Matplotlib 기본 Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False # x,y축 레이블 - 깨지는 현상방지
```

### 1. 라인플롯

- 그래프를 꾸밀 수 있는 것들
    - plt.title()
    - plt.xlabel()    plt.ylabel()
    - plt.legend()
    - plt.grid()   : 격자무늬
    - plt.xlim()      plt.ylim()        x,y축 범위
    - plt.xticks()    plt.yticks()      x,y축 
    - plt.show()

```python
plt.figure(figsize=(5, 3))
plt.plot([0,1,2,3], [0,1,4,9])
plt.title('라인 플롯'), plt.grid()
plt.show()
```

```python
plt.figure(figsize=(5, 3))
plt.plot([0,1,2,3], [0,1,4,9])
plt.title('라인 플롯'), plt.grid();     # plt.show() 생략하면 마지막에 ;을 붙여줄것
```

- 스타일 지정

```python
plt.figure(figsize=(5, 3))
plt.plot( [0,1,4,9], 'rs--')               # X축 [0,1,2,3] 생략됨
# 'rs--' : color: r(red), marker: s(square), linestyle: --
plt.title("'rs--' 스타일의 플롯"), plt.grid()
plt.show()
```

```python
plt.figure(figsize=(5, 3))
plt.plot( [0,1,4,9], 'mo-.')  
plt.plot( [9,4,1,0], 'g^:')  # ^ maker : triangle
plt.title("'mo-.', 'g^:'  스타일의 플롯"), plt.grid()
plt.show()
```

- 그림 범위

```python
plt.title("x축, y축의 범위 설정")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.xlim(0, 50), plt.ylim(-10, 30), plt.grid()      ## <-- 해당줄은 자주 사용함
plt.show()
```

- 곡선 그리기

```python
X = np.linspace(0, 2*np.pi, 200)
Ysin, Ycos = np.sin(X), np.cos(X)
plt.plot(X, Ysin), plt.plot(X, Ycos)
plt.title('Sine, Cosine 그래프'), plt.grid()
plt.show()
```

- Tick 설정

```python
plt.plot(X, Ysin), plt.plot(X, Ycos)
plt.title('Tick 설정'), plt.grid()
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
plt.yticks([-1, -0.707, 0, 0.707, 1])
plt.show()
```

```python
plt.plot(X, Ysin), plt.plot(X, Ycos)
plt.title('LaTeX, 문자열로 tick label 정의'), plt.grid()
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])    # LaTeX
plt.yticks([-1, -0.707, 0, 0.707, 1],
           [r'$-1$', r'$-\sqrt{2}/2$', r'$0$', r'$\sqrt{2}/2$', r'$1$'])

plt.yticks([-1, -0.707, 0, 0.707, 1])
plt.show()
```
