## 서울시 CCTV - 인구 시각화

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('../02.Pandas/data/서울시 구별 CCTV 인구 현황.csv', index_col='구별')
df.head()
```

### 4. 인구수와 CCTV 댓수

- 산점도

```python
plt.figure(figsize=(6,6))
plt.scatter(df['인구수'], df['CCTV댓수'])
plt.title('인구수와 CCTV 댓수의 관계')
plt.grid(), plt.xlabel('인구수'), plt.ylabel('CCTV댓수')
plt.show()
```

- 회귀선  

```python
fp1 = np.polyfit(df['인구수'], df['CCTV댓수'], 1)        #poly : 1차식 2차식 3차식
fp1                     # [기울기, 절편]
```

```python
fx = np.array([100000, 700000])     # 표에서 인구수 최소 값~최대수...
fy = fx * fp1[0] + fp1[1]       #fx *기울기 + 절편
fy
```

```python
f1 = np.poly1d(fp1)         # X값을 주면 Y값을 구해주는 함수
fy = f1(fx)
fy
```

```python
plt.figure(figsize=(6,6))
plt.scatter(df['인구수'], df['CCTV댓수'])
plt.plot(fx, fy, 'g--', lw=3)
plt.title('인구수와 CCTV 댓수의 관계', fontsize=15)
plt.grid(), plt.xlabel('인구수'), plt.ylabel('CCTV댓수')
plt.show()
```

- 오차의 절대값

```python
df['오차'] = abs(df['CCTV댓수'] - f1(df['인구수'])).round(1)
df.head(3)
```

```python
# 오차의 내림차순으로 정렬
df.sort_values(by='오차', ascending=False).head()
```

```python
# 정렬된 데이터에서 Top 10만 골라내기
df_sort = df.sort_values(by='오차', ascending=False).head(10)
df_sort
```

- 최종 그림

```python
plt.figure(figsize=(12, 9))
plt.scatter(df['인구수'], df['CCTV댓수'], c= df['오차'])     # c : color의 약자
plt.plot(fx, fy, 'g--', lw=3)


# plt.text(df_sort['인구수'][0]+8000, df_sort['CCTV댓수'][0]-50, df_sort.index[0], fontsize=15)
# ↑ 강남구 점에만 글자 표시해보기 = >  X좌표 : df_sort['인구수'][0] = 강남구, Y자표:df_sort['CCTV댓수'][0]
# 위에꺼를 반복문을 써서 모든 점에
for i in range(10):
    if i == 8:
        continue
    plt.text(df_sort['인구수'][i]+8000, df_sort['CCTV댓수'][i]-50,
             df_sort.index[i], fontsize=15)
plt.text(df_sort['인구수'][8]-5000, df_sort['CCTV댓수'][8]-200,
             df_sort.index[8], fontsize=15)

plt.title('인구수와 CCTV 댓수의 관계', fontsize=20)
plt.grid(), plt.xlabel('인구수'), plt.ylabel('CCTV댓수')
plt.colorbar()
plt.show()
```
