## 서울시 CCTV - 인구 시각화

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('../02.Pandas/data/서울시 구별 CCTV 인구현황.csv', index_col='구별')
df.head()
```

##### 1. 자치구별 CCTV 설치댓수

```python
df.CCTV댓수.plot(kind= 'barh')
plt.title('자치구별 CCTV 설치댓수');
```

```python
# CCTV 댓수로 정렬하고, plt.barh()로 
sc = df['CCTV댓수'].sort_values()
plt.barh(sc.index, sc.values)
plt.title('서울시 구별 CCTV 설치댓수 정렬완료');
```

```python
# Seaborn으로 그리기
sns.barplot(data=df, y=df.index, x='CCTV댓수')
plt.title('서울시 구별 CCTV 설치댓수 Seaborn')
```

```python
sns.barplot(y=sc.index, x=sc.values);
```

```python
scr = df['CCTV댓수'].sort_values(ascending=False)
sns.barplot(y=sc.index, x=sc.values)
plt.title('서울시 구별 CCTV 설치댓수 Seaborn');
```
